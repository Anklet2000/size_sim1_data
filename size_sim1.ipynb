```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.rcParams['font.family'] = "MS Gothic"
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
```

```python
df=pd.read_csv('sim1data.txt', sep=' ', engine='python', header=None)
```

```python
xs = np.array(df[0])
ys = np.array(df[1])
```

```python
plt.plot(xs, ys, color="black")
plt.xlabel("試行回数(回)")
plt.ylabel("系のエネルギー")
plt.xlim(0,10500)
plt.show()
```
